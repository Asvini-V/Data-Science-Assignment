# Q7
Test the Analysis of Variance between etest_p and mba_p at signifance level 5%.(Make decision using Hypothesis Testing)

```python
from scipy.stats import f_oneway
f_stat, p_value = f_oneway(dataset['etest_p'], dataset['mba_p'])
print("F-Statistic:", f_stat)
print("P-Value:", p_value)
```

```python
# One-Way ANOVA is used to test whether there is a significant difference between etest_p and mba_p.

# H₀: There is no significant difference between etest_p and mba_p.
# H₁: There is a significant difference between etest_p and mba_p.

# From the result, the p-value is 4.67e-21.
# Since the p-value is less than 0.05, the null hypothesis is rejected.
# Hence, there is a significant difference between etest_p and mba_p
```

# Q8
Test the similarity between the degree_t(Sci&Tech) and specialisation( Mkt&HR) with respect to salary at significance level of 5%.
(Make decision using Hypothesis Testing)

```python
from scipy.stats import ttest_ind
```

```python
group1 = dataset[dataset['degree_t'] == 'Sci&Tech']['salary']
group2 = dataset[dataset['specialisation'] == 'Mkt&HR']['salary']
```

```python
t_stat, p_value = ttest_ind(group1, group2)

print("T-Statistic:", t_stat)
print("P-Value:", p_value)
```

```python
# Independent T-test is used to compare the salary between students from Sci&Tech degree and Mkt&HR specialization.

# H₀: There is no significant difference in salary.
# H₁: There is a significant difference in salary.

# From the result, the p-value is 0.0079.
# Since the p-value is less than 0.05, the null hypothesis is rejected.

# Hence, there is a significant difference in salary between the two groups.
```

# Q9
Convert the normal distribution to standard normal distribution for salary column

```python
def stdNBgraph(dataset):
    # Converted to Standard Normal Distribution
    
    import seaborn as sns
    import numpy as np
    
    mean = dataset.mean()
    std = dataset.std()

    values = [i for i in dataset]

    z_score = [((j - mean) / std) for j in values]

    sns.distplot(z_score, kde=True)

    print("Mean of Z-score:", sum(z_score) / len(z_score))
    print("Std of Z-score:", np.std(z_score))
```

```python
stdNBgraph(dataset['salary'])
```

```python
# The salary column is converted into standard normal distribution using Z-score. The mean of the transformed data is approximately 0 and the 
# standard deviation is approximately 1. The distribution plot shows that the data is centered around zero, indicating successful standardization.
```

# Q10
What is the probability Density Function of the salary range from 700000 to 900000?

```python
def get_pdf_probability(dataset,startrange,endrange):
    from matplotlib import pyplot
    from scipy.stats import norm
    import seaborn as sns
    ax = sns.distplot(dataset,kde=True,kde_kws={'color':'blue'},color='Green')
    pyplot.axvline(startrange,color='Red')
    pyplot.axvline(endrange,color='Red')
    # generate a sample
    sample = dataset
    # calculate parameters
    sample_mean =sample.mean()
    sample_std = sample.std()
    print('Mean=%.3f, Standard Deviation=%.3f' % (sample_mean, sample_std))
    # define the distribution
    dist = norm(sample_mean, sample_std)
    
    # sample probabilities for a range of outcomes
    values = [value for value in range(startrange, endrange)]
    probabilities = [dist.pdf(value) for value in values]    
    prob=sum(probabilities)
    print("The area between range({},{}):{}".format(startrange,endrange,sum(probabilities)))
    return prob
```

```python
get_pdf_probability(dataset['salary'], 700000, 900000)
```

```python
# The probability density function is used to calculate the probability of salary lying between 700000 and 900000.

#  From the result, the probability value is 0.000597.

#  This indicates that the likelihood of students getting salary in this range is extremely low.
```

# Q11
Test the similarity between the degree_t(Sci&Tech)with respect to
etest_p and mba_p at significance level of 5%.(Make decision using Hypothesis
Testing)

```python
from scipy.stats import ttest_rel
```

```python
group = dataset[dataset['degree_t'] == 'Sci&Tech']
```

```python
t_stat, p_value = ttest_rel(group['etest_p'], group['mba_p'])

print("T-Statistic:", t_stat)
print("P-Value:", p_value)
```

```python
# Paired T-test is used to compare etest_p and mba_p for students from Sci&Tech degree.

# H₀: There is no significant difference between etest_p and mba_p.
# H₁: There is a significant difference between etest_p and mba_p.

# From the result, the p-value is 5.51e-06.

# Since the p-value is less than 0.05, the null hypothesis is rejected.

# Hence, there is a significant difference between etest_p and mba_p for Sci&Tech students.
```
